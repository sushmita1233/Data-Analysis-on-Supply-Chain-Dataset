```python
# code snippet 18
# filtering dataframe to include only one rows where sellercountry is "Italy"
filter_df = df[df['SellerCountry']=="Italy"]
filter_df
```

```python
# code snippet 19
# filtering the dataframe to include only one row where SellerCountry is "US" and price>100
filtered_df = df[(df['SellerCountry']=="US") & (df['Price']>100)]
filtered_df
```

```python
# code snippet 20
# Filtering data for specific categories is either "Beauty" or "Health"
filter_categories = df[df['Category'].isin(['Beauty', 'Health'])]
filter_categories
```

```python
# code snippet 21
# filtering data for shipping anf feedback conditions where shipiingCost is not more than 10 and SellerFeedback is freater than 95 in descending order

filter_shipping = df[(df['ShippingCost']<=10) & (df['SellerFeedback']>=95)].sort_values(by='SellerFeedback', ascending=False)
filter_shipping
```

```python
# code snippet 22
# renaming the SellerFeedback couln to review rating in the dataframe
df = df.rename(columns={'SellerFeedback': 'ReviewRating'})
df
```

```python
# code snippet 23
# calculating totalsales for category
home_improvement_sales = df[df['Category'] == 'Home Improvement']
total_sales = (home_improvement_sales['Quantity']*home_improvement_sales["Price"]).sum()
total_sales
```

```python
# code snippet 24
# calculating avg Shipping cost in Outdoor Category
outdoor_category = df[df['Category'] == 'Outdoors']
avg_shipping_cost = outdoor_category['ShippingCost'].mean()
avg_shipping_cost
```

```python
# code snippet 25
# counting and summing price for beauty products from china in Beauty category from seller "china "
beauty_products = df[(df['Category'] == 'Beauty') & (df['SellerCountry'] == 'China')]
total_price = beauty_products['Price'].sum()
count_products = beauty_products['Price'].count()
total_price
```

```python
# code snippet 27
# adding new column  called"BuyerCountry" to the dataframe
df['BuyerCountry'] = 'None'
df
```

```python
# code snippet 28
# Calculating avg price for each category in the electronic
avg_price_by_category = df[df['Category'] == 'Electronics'].groupby('SubCategory')['Price'].mean()
avg_price_by_category
```

```python
# code snippet 29
# filtering delivery time and quantity where delivery time is less than or equal to 5 days and quantity is greater than 10
filter_delivery = df[(df['DeliveryTime']<=5) & (df['Quantity']>10)]
filter_delivery
```

```python
# code snippet 30
# calculating the maximum price amoung all the products
max_price = df['Price'].max()
max_price
```

```python
# code snippet 31
# calculating total sales per seller  and grouped by SellerID
# creating new column 'ToalSales' multiplying price and quantity
df["TotalSales"] = df["Price"] * df["Quantity"]
# groupby sellerId and sum the totalsales for each seller
total_sales_per_seller = df.groupby('SellerId')['TotalSales'].sum()
total_sales_per_seller
```
